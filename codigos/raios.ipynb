```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('/content/powerespec.xls')
df['Power (dBm)'] = df['Power (dBm)']/1000
df['Phase'] = df['Phase'].str.split().str[1]
df['Power (dBm)']
```

```python
df.columns
```

```python
df['Distance (m)'] = df['Time (s)'] * 3*10**8
```

```python
len(df['receiver'].unique())
```

```python
plt.figure(figsize=(12, 8))


# Gerando o grÃ¡fico de potÃªncia em funÃ§Ã£o do tempo
power = np.array(df[df['receiver']==1][['Power (dBm)']])
time = np.array(df[df['receiver']==1][['Time (s)']])

#plt.ylim((power.max(), power.min()))
plt.scatter(time,power)
plt.vlines(time,-100,power)
plt.xlabel('Tempo (s)')
plt.ylabel('PotÃªncia (dBm)')
plt.title('GrÃ¡fico de PotÃªncia em FunÃ§Ã£o do Tempo')
plt.grid(True)
plt.axis([min(time)-0.5e-8, max(time)+0.5e-7, min(power)-10, max(power)+10])
```
